```python
Race_index = Race_df.reset_index()
Race_index
```

```python
Race_index.replace("Unknown/Unavailible", "Unknown", inplace=True) 
Race_index
```

```python
Race_index.replace("Unknown/Unavailable", "Unknown", inplace=True) 
Race_index
```

```python
Race_index.replace("Unknown/Unavailable", "Unknown", inplace=True) 
Race_index
```

```python
Race_list = Race_index.groupby(['Race'], sort=True)['count'].sum()    
Race_list
```

```python
Race_list_df = pd.DataFrame(Race_list)
Race_list_df
```

```python
Race_index2 = Race_list_df.reset_index()
Race_index2 
```

```python
Race_index2.sort_values(by = 'count', ascending = False)
Race_Counts = Race_index2.sort_values(by = 'count', ascending = False)
Race_Counts
```
